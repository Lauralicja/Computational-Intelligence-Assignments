# Clustering comparison

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
sb.set()

data = pd.read_csv('AllBooks_baseline_DTM_Unlabelled.csv')
data
```

# Reading the data

```python
df = pd.DataFrame(data)
num_words = len(df.columns)
num_sentences = len(df.rows)
```

```python
data_arr_values = np.zeros(num_sentences)
data_arr_names = np.zeros(num_words).astype(str)
names = df.columns.astype(str)
df.iloc[1, :].max()
df.dropna()
```

```python
for row in range(num_sentences):
    print(row)
    data_arr_values[row] = df.iloc[row, :].max().astype(int)
    for cols in range(num_words):
        if df.iloc[row, cols] == df.iloc[row, :].max():
            data_arr_names[row] = names[cols]
```

# Plotting the data

```python
time_arr_vals = np.linspace(0, data_arr_values.size-1, data_arr_values.size, endpoint=True)
plt.scatter(time_arr_vals, data_arr_values)
data_arr_vals_list = data_arr_values.tolist()
```

```python
time_arr_names = np.linspace(0, data_arr_names.size-1, data_arr_names.size, endpoint=True)
plt.figure(figsize=(10,30))
plt.scatter(time_arr_names, data_arr_names)
```

```python
# removing corrupted data
data_arr_names_clear = np.delete(data_arr_names, np.where(data_arr_names == '0.0'))
```

Plot of the words per sentence.

```python
plt.plot(data_arr_names_clear)
```

Plot of the most common words

```python
plt.scatter(data_arr_values, data_arr_names_clear)
```

# Implementing the clustering models

```python
from sklearn.cluster import DBSCAN, KMeans

kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(data_arr_values.reshape(-1,1))

dbscan = DBSCAN(eps=0.3, min_samples=4)
dbscan.fit(data_arr_values.reshape(-1,1))
```

```python
predicted_kmeans = kmeans.fit_predict(data_arr_values.reshape(-1,1))
predicted_dbscan = dbscan.fit_predict(data_arr_values.reshape(-1,1))

print(predicted_kmeans)
print(predicted_dbscan)
```

## Printing the most common words for cluster with a certain number 

```python
print(np.unique(data_arr_names_clear[np.where(predicted_kmeans == 3)]))
```

```python
print(np.unique(data_arr_names_clear[np.where(predicted_dbscan == 20)]))
```

## Plot of KMeans clusters

```python
plt.figure(figsize=(10,30))
plt.scatter(data_arr_values, data_arr_names_clear, c=predicted_kmeans, cmap='rainbow')
```

## Plot of DBScan clusters

```python
plt.figure(figsize=(10,30))
plt.scatter(data_arr_values, data_arr_names_clear, c=predicted_dbscan, cmap='rainbow')
```
